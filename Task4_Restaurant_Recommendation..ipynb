```python
# 1. Import Required Libraries
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 2. Load the Dataset
df = pd.read_csv("Dataset.csv")

# 3. Select Necessary Columns
df = df[['Restaurant Name', 'Cuisines', 'Average Cost for two', 'Aggregate rating']]
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)

# 4. Preprocess Data
# Lowercase and clean cuisines
df['Cuisines'] = df['Cuisines'].str.lower().str.strip()
# Combine cuisine and cost for content features
df['combined_features'] = df['Cuisines'] + " " + df['Average Cost for two'].astype(int).astype(str)

# 5. Feature Extraction using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english')
feature_matrix = vectorizer.fit_transform(df['combined_features'])

# 6. Compute Similarity Matrix
similarity_matrix = cosine_similarity(feature_matrix)

# 7. Recommendation Function
def recommend_restaurants(cuisine, max_cost, top_n=5):
    """
    Recommend restaurants based on user's cuisine preference and budget.

    Parameters:
        cuisine (str): Preferred cuisine (e.g., 'Italian')
        max_cost (int): Maximum cost for two people
        top_n (int): Number of recommendations to return

    Returns:
        DataFrame: Recommended restaurants
    """
    # Create input feature like in dataset
    user_input = cuisine.lower().strip() + " " + str(int(max_cost))
    input_vector = vectorizer.transform([user_input])

    # Calculate similarity scores
    similarity_scores = cosine_similarity(input_vector, feature_matrix).flatten()

    # Get top indices
    top_indices = similarity_scores.argsort()[::-1]

    # Filter by budget
    recommended = df.iloc[top_indices]
    recommended = recommended[recommended['Average Cost for two'] <= max_cost]

    return recommended[['Restaurant Name', 'Cuisines', 'Average Cost for two', 'Aggregate rating']].head(top_n)

# 8. Test the Recommendation System
# Example: Recommend South Indian food under ₹300
recommendations = recommend_restaurants("South Indian", 300)
print("Top Recommendations:\n")
print(recommendations)
```
